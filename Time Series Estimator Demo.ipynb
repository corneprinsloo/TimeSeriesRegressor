# Stock Market Prediction with Time Series Estimators

---

The TimeSeriesEstimator is a wrapper object for sci-kit learn estimators. It is designed to aid in predicting sequence to sequence data, one example of which is predicting the next term in the original sequence. This object is a Meta-Estimator which means that it takes in another sklearn eatimator as a parameter and uses that to model the data. The following notebook is intended to show the ease of use of the object and its power for predicting the stock market.

---

# Data Visualization

First lets load and process the data! The data contains the adjusted closing prices of all s&p 500 stocks which have prices in the range from 2014-1-1 to the current date.


```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from TimeSeriesEstimator import TimeSeriesRegressor, time_series_split
from utils import datasets

X = datasets('sp500')
y = X['AAPL']
X_train, X_test = time_series_split(X)
y_train, y_test = time_series_split(y)

y.plot()
plt.title("Adjusted Close of APPLE")
plt.show()
```

# Basic Usage

Now that we have some idea of what our data looks like, let's fit it with a time series regressor (TSR). The TSR takes in a linear model, and the parameter n_prev, which sets how far back in the past the estimator should look when fitting data. 

```python
from sklearn.linear_model import LinearRegression

n_prev=2
linear_model = LinearRegression()
linear_tsr = TimeSeriesRegressor(linear_model, n_prev=n_prev)
linear_tsr.fit(X_train, y_train)
pred_train = linear_tsr.predict(X_train) #outputs a numpy array of length: len(X_train)-n_prev
pred_test = linear_tsr.predict(X_test)
```

Now to visualize the results!

```python
def train_test_plot(pred_train,y_train,pred_test,y_test,n_prev, title, n_ahead=1):
    plt.subplot(1,2,1)
    plt.plot(pred_train[:], 'r', label="Predicted")
    plt.plot(y_train[n_prev-1+n_ahead:], 'b--', label="Actual") 
    #nprev: because the first predicted point needed n_prev steps of data
    plt.title("Training performance of "+title)
    plt.legend(loc='lower right')

    plt.subplot(1,2,2)
    plt.plot(pred_test[:], 'r', label="Predicted")
    plt.plot(y_test[n_prev-1+n_ahead:], 'b--', label="Actual") 
    #nprev: because the first predicted point needed n_prev steps of data
    plt.title("Testing performance of " +title)
    plt.legend(loc='lower left')

    plt.gcf().set_size_inches(15,6)
    plt.show()

train_test_plot(pred_train,y_train,pred_test,y_test,n_prev,'linear TSR on APPLE')
```

Woah woah woah, that doesent look that good! What happened? We have to remember that the dataset contains 500 variables and were just trying to predict AAPL. We are in for some serious overfitting at this rate. Lets improve our model by adding an L1 and L2 weight penalty on the fit. We can accomplish this by using a generalization of the Lasso called the ElasticNet estimator instead of the standard linear regression.

```python
from sklearn.linear_model import ElasticNet

n_prev=1
elastic = ElasticNet()
elastic_tsr = TimeSeriesRegressor(elastic, n_prev=n_prev,alpha=1, max_iter=5000) 
# Note that ElasticNet parameters can be input in the ElasticNet() or the TimeSeriesRegressor()
# This allows grid search and other meta estimation techniques to access the parameters of the base_estimator
elastic_tsr.fit(X_train, y_train)
pred_train_2 = elastic_tsr.predict(X_train) #outputs a numpy array of length: len(X_train)-n_prev
pred_test_2 = elastic_tsr.predict(X_test)

train_test_plot(pred_train_2,y_train,pred_test_2,y_test,n_prev,'ElasticNet TSR on APPLE')
```

# Predicting Multiple Days into the Future with n_ahead

```python
from sklearn.linear_model import ElasticNet

n_prev=2; n_ahead=10
lasso = ElasticNet(alpha=1, l1_ratio=1)
lasso_tsr = TimeSeriesRegressor(lasso, n_prev=n_prev,n_ahead=n_ahead,alpha=1) 
# Note that ElasticNet parameters can be input in the ElasticNet() or the TimeSeriesRegressor()
# This allows grid search and other meta estimation techniques to access the parameters of the base_estimator
lasso_tsr.fit(X_train, y_train)
pred_train_2 = lasso_tsr.predict(X_train) #outputs a numpy array of length: len(X_train)-n_prev
pred_test_2 = lasso_tsr.predict(X_test)

train_test_plot(pred_train_2,y_train,pred_test_2,y_test,n_prev,'ElasticNet TSR on APPLE',n_ahead)
```

# Forecasting Arbitrarily Many Days Ahead

Sometimes it is desireable to be able to predict arbitrarily far into the future. To do this, a model needs to output exactly what it takes in as input, so that the output can be fed back into the model and predict another day into the future. The TSR class supports this functionality through the forecast method which takes the training data and a number of steps to predict and moves the model forward in time. The model needs to be an autocorrelation model (i.e no target sequence Y) and needs to only predict 1 day ahead (i.e n_ahead = 1). It is generalized to handle any setting of n_prev. 

Lets try forecasting the entire s&p500 ~93 days ahead (the length of the testing dataset) with a linear regressor!

```python
X = datasets('sp500')
names = list(X.columns.values)
X_train, X_test = time_series_split(X)

n_prev = 3
tsr = TimeSeriesRegressor(LinearRegression(), n_prev=n_prev)
tsr.fit(X_train)
fc = tsr.forecast(X_train, len(X_test))
def changes(X):
    return np.array([X[-1, i] - X[0, i] for i in range(X.shape[1])])

plt.subplot(1,3,1)
stock=3
plt.plot(fc[:,stock],'r',label='Forecast of '+names[stock])
plt.plot(X_test[:,stock],'b',label='Actual price of '+names[stock])
plt.ylabel('Price')
plt.xlabel('Day')
plt.legend(loc='upper left')

plt.subplot(1,3,2)
stock=11
plt.plot(fc[:,stock],'r',label='Forecast of '+names[stock])
plt.plot(X_test[:,stock],'b',label='Actual price of '+names[stock])
plt.ylabel('Price')
plt.xlabel('Day')
plt.legend(loc='upper left')

plt.subplot(1,3,3)
X_test_change = changes(np.log(X_test))
fc_change = changes(np.log(fc))
x = np.linspace(min(X_test_change), max(X_test_change))
plt.plot(x,x,'b',label='y = x')
plt.plot(X_test_change, fc_change, 'ro',label='data', alpha=.3)
plt.title('Predicted vs. Actual Changes in Price')
plt.xlabel('Actual price change in log(dollars)')
plt.ylabel('Forecasted price change in log(dollars)')
plt.legend(loc='upper left')

plt.gcf().set_size_inches(16,5)
plt.show()
```

# Investigating How Accuracy Varies with Forecast Length

Lets investigate how well we can predict the changes in prices as we increase the forecast length. A correlation above 0 means that we are doing above random for predicting the changes in stock price

```python
X = datasets('sp500')
names = list(X.columns.values)
X_train, X_test = time_series_split(X)

for n_prev in range(1, 5):
    tsr = TimeSeriesRegressor(LinearRegression(), n_prev=n_prev)
    tsr.fit(X_train)
    fc = tsr.forecast(X_train, len(X_test))

    def changes(X, start=0,end=-1):
        return np.array([X[end, i] - X[start, i] for i in range(X.shape[1])])

    n_aheads = range(1,len(X_test),1)
    X_test_changes = [changes(X_test,end=n_ahead) for n_ahead in n_aheads]
    fc_changes = [changes(fc,end=n_ahead) for n_ahead in n_aheads]
    cors = [np.corrcoef(X_test_changes[i],fc_changes[i])[0,1] for i in range(len(n_aheads))]
    plt.plot(n_aheads,cors, label="n_prev={}".format(n_prev))

plt.xlabel('Length of time forecasted')
plt.ylabel('Correlation of price changes')
plt.legend()
plt.show()
```

Predicting ~30 days ahead seems to be a sweet spot for this problem.

# Using GridSearchCV and other Meta-Parameter Estimation Methods

The time series estimator class works with any sklearn model including Pipelines and other composite and Meta estimators. It also works with grid search, provided that one uses the time_series_cv, or cascade_cv function to create the cross validation iterator. 

The time_series_cv function works by splitting your data into n_folds nonoverlapping temporally contiguous sections. This is good is one has a large ammount of data. If one would like to be more data efficient use cascade_cv which takes in a data_size fraction (default .8) and a number of folds. cascade_cv slides the .8 * len(data) window along the dataset so that after n_folds the window stretches to the end of the dataset.

```python
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import KFold
from TimeSeriesEstimator import cascade_cv

X = datasets('sp500')
y = X['AAPL']
X_train, X_test = time_series_split(X)
y_train, y_test = time_series_split(y)

n_prev = 3
tsr = TimeSeriesRegressor(ElasticNet(max_iter=1000), n_prev=n_prev)

param_grid = [{'alpha': [.01, .05, .1, .5, 1, 5, 10], 
               'l1_ratio': [0,.25,0.5,.75,1]}]
cv = cascade_cv(len(X_train), n_folds=5)
grid = GridSearchCV(tsr, param_grid, cv=cv)
grid.fit(X_train, y_train)
pred_train_3 = grid.predict(X_train)  # outputs a numpy array of length: len(X_train)-n_prev
pred_test_3 = grid.predict(X_test)

#print(grid.grid_scores_)
print(grid.best_params_)
train_test_plot(pred_train_3, y_train, pred_test_3, y_test, n_prev, 'Optimized Lasso TSR on APPLE')
```

# The Mechanics of the TSR

Now seems like a great time to discuss what the TSR is actually doing underneath the hood. The TSR works by taking in a single (X) or two datasets (X,Y) of equal length. In the single dataset case, the TSR assumes you would like to predict the next element in the dataset using the previous elements. In either case it forms a dataset by taking the previous n timesteps and flattening them into a vector.

# Tuning the Window Size Parameter

 Lets levearage this generality to easily perform model selection. Lets see how window size effects the performance of the linear model. This time lets make the problem harder! Instead of trying to capture the effect of the stock market on the price of Apple, lets see the stock market's effect on the price of every stock in the s&p 500! This makes the problem exponentially harder!

```python
from utils import safe_shape, mse

X = datasets('sp500')
X_train, X_test = time_series_split(X) 
#Note that we only need one dataset because the TSR knows we want to predict the 
#entire state of the market at a future time step


#Lets investigate 10 settings of n_prevs
n_prevs = range(1, 10)
mses_train = np.empty((len(n_prevs), safe_shape(X_train, 1)))
mses_test =  np.empty((len(n_prevs), safe_shape(X_train, 1)))
mses_forecast =  np.empty((len(n_prevs), safe_shape(X_train, 1)))

for i, n_prev in enumerate(n_prevs):
    tsr = TimeSeriesRegressor(LinearRegression(), n_prev=n_prev)
    tsr.fit(X_train)
    mses_train[i, :] = mse(tsr.predict(X_train), X_train[n_prev:])
    mses_test[i, :] = mse(tsr.predict(X_test), X_test[n_prev:])
    mses_forecast[i, :] = mse(tsr.forecast(X_train,len(X_test)), X_test)

plt.subplot(1,2,1)
plt.boxplot(np.log(np.transpose(mses_test)))
#plt.yscale('log')
plt.title("TSR prediction over the entire s&p500")
plt.ylabel("Testing log(MSE)s of the 500 stocks")
plt.xlabel("Setting of n_prev")

plt.subplot(1,2,2)
plt.boxplot(np.log(np.transpose(mses_forecast)))
#plt.yscale('log')
plt.title("TSR forecast over the entire s&p500")
plt.ylabel("Testing log(MSE)s of the 500 stocks")
plt.xlabel("Setting of n_prev")

plt.gcf().set_size_inches(10,5)
plt.show()
```

Hence, a setting of n_prev=2 should be sufficient to imporove the performance of the linear model. Lets now try different types of models to see its sucess.

# Model Selection on the Entire S&P 500

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.kernel_ridge import KernelRidge
from sklearn.base import clone
from sklearn.linear_model import ElasticNet,Lasso

X = datasets('sp500')
X_train, X_test = time_series_split(X) 
#Note that we only need one dataset because the TSR knows we want to predict the 
#entire state of the market at a future time step

models= [LinearRegression(), DecisionTreeRegressor(),Lasso(),ElasticNet(), KernelRidge(), KernelRidge(kernel='rbf')]
names= ['Linear', 'DecisionTree','Lasso','ElasticNet','linear-KR','rbf-KR']
mses_train = np.empty((len(models), safe_shape(X_train, 1)))
mses_test =  np.empty((len(models), safe_shape(X_train, 1)))
mses_forecast =  np.empty((len(models), safe_shape(X_train, 1)))


n_prev=2
for i, model in enumerate(models):
    tsr = TimeSeriesRegressor(clone(model), n_prev=n_prev)
    tsr.fit(X_train)
    mses_train[i, :] = mse(tsr.predict(X_train), X_train[n_prev:])
    mses_test[i, :] = mse(tsr.predict(X_test), X_test[n_prev:])
    mses_forecast[i, :] = mse(tsr.forecast(X_train,len(X_test)), X_test)

plt.subplot(1,2,1)
plt.boxplot(np.log(np.transpose(mses_test)))
plt.title("TSR prediction over the entire s&p500")
plt.ylabel("Testing log(MSE)s of the 500 stocks")
plt.xticks(range(1,len(names)+1), names)

plt.subplot(1,2,2)
plt.boxplot(np.log(np.transpose(mses_forecast)))
plt.title("TSR forecast over the entire s&p500")
plt.ylabel("Testing log(MSE)s of the 500 stocks")
plt.xticks(range(1,len(names)+1), names)

plt.gcf().set_size_inches(14,5)
plt.show()
```
